```python
from pyorbits import calculate_orbit

x_final, y_final = calculate_orbit(time_max = 300*365*24*60*60, dt = 60*10)
print(f"Calculated {len(x_final)} x {len(y_final)} points")
```

```python
import matplotlib.pyplot as plt

plt.figure(1)
plt.subplot(1,2,1)
plt.plot(x_final,y_final,linewidth = 2)
plt.grid(linestyle=(0, (10, 5)), color='gray', linewidth=0.5)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('x [Au]', fontsize=15)
plt.ylabel('y [Au]', fontsize=15)
plt.axis('equal')
```
